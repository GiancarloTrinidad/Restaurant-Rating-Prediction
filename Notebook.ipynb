```python
X = df[["Average Cost for two", "Has Table booking", "Has Online delivery", "Price range"]]
y = df[["Aggregate rating"]]
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
X["Has Table booking"] = labelencoder.fit_transform(X["Has Table booking"])
```

```python
X
```

```python
X["Has Online delivery"] = labelencoder.fit_transform(X["Has Online delivery"])
```

```python
X
```

For table booking and online delivery:
0 = No,
1 = Yes

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

Model Building

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test, predictions)))
    print("Root mean squared error on model is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_predictions = lr.predict(X_test)
modelresults(lr_predictions)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
svrmodel = SVR()
svr_param_grid = {"C" : [0.01,0.05], "kernel" : ["linear","poly"], "degree": [2]}
svr_grid_search = GridSearchCV(svrmodel, svr_param_grid)
svr_grid_search.fit(X_train,y_train)
```

```python
svr_predictons = svr_grid_search.predict(X_test)
modelresults(svr_predictons)
```

```python
from sklearn.tree import DecisionTreeRegressor
treemodel = DecisionTreeRegressor()
tree_param_grid = {"max_depth" : [10,30], "min_samples_leaf" : [1,2], "min_samples_split" : [2,5]}
tree_grid_search = GridSearchCV(estimator = treemodel, param_grid = tree_param_grid)
tree_grid_search.fit(X_train,y_train)
```

```python
tree_predictions = tree_grid_search.predict(X_test)
modelresults(tree_predictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
rfr_param_grid = {"max_depth" : [5,15], "n_estimators" : [2,5,10]}
rfr_grid_search = GridSearchCV(rfrmodel, rfr_param_grid)
rfr_grid_search.fit(X_train,y_train)
```

```python
rfr_predictions = rfr_grid_search.predict(X_test)
modelresults(rfr_predictions)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors" : [25,26,27,28]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
knn_grid_search.fit(X_train,y_train)
```

```python
knn_predictions = knn_grid_search.predict(X_test)
modelresults(knn_predictions)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada = AdaBoostRegressor()
ada_param_grid = {"n_estimators" : [50,100,200], "learning_rate" : [0.1,0.5,1]}
ada_grid_search = GridSearchCV(ada,ada_param_grid)
ada_grid_search.fit(X_train,y_train)
```

```python
ada_predictions = ada_grid_search.predict(X_test)
modelresults(ada_predictions)
```
